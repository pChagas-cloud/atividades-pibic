```python
import tensorflow as tf
import numpy as np
```

```python
#Setting up filenames (X)

filenames = []

for i in range(10):
    name = f"A{i+1}.png"
    filenames.append(name)
    
for i in range(10):
    name = f"B{i+1}.png"
    filenames.append(name)
```

```python
#Setting up labels (Y)

labels = []

for i in filenames:
    if "A" in i:
        labels.append(0)
    elif "B" in i:
        labels.append(1)
    
labels = np.array(labels, dtype=np.uint8)
```

```python
#create data structure

dataset = tf.data.Dataset.from_tensor_slices((filenames, labels))
```
